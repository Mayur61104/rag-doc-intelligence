LOADING PDF AND CHUNKING

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
def load_and_chunk(pdf_path):
  loaded = PyPDFLoader(pdf_path)
  pages = loaded.load()
  print(len(pages),"pages loaded")
  #split into chunks
  splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,    # each chunk = max 512 characters
        chunk_overlap=100,  # 64 chars shared between neighbouring chunks
        separators=["\n\n", "\n", ". ", " ", ""]
        # tries to split on blank lines first, then single lines,
        # then sentences, then words — in that priority order
    )
  chunks = splitter.split_documents(pages)
  print(len(chunks),"chunks created")
  return chunks
  print(f"Created {len(chunks)} chunks from {len(pages)} pages")
  print(f"Average chunk size: {sum(len(c.page_content) for c in chunks)//len(chunks)} chars")

  # Preview first chunk so you can see what it looks like
  print(f"\nFirst chunk preview:")
  print("-" * 40)
  print(chunks[0].page_content[:300])
  print("-" * 40)

```

```python
chunks = load_and_chunk(file_name)
```

BUILDING VECTORSTORE (EMBEDDINGS)

```python
# Updated imports for current LangChain versions
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from sentence_transformers import SentenceTransformer
```

```python
def build_vector_store(chunks):
  print('loading embeddings model......')
  embeddings = HuggingFaceEmbeddings(
    model_name="BAAI/bge-base-en-v1.5",
    model_kwargs={"device": "cpu"},
    encode_kwargs={"normalize_embeddings": True}
)
  print('embeddings model loaded')
  print(f"\nEmbedding {len(chunks)} chunks...")
  print("(Converting each chunk to a list of 384 numbers)")
  #faiss to store chunks and embeddings
  vectorstore = FAISS.from_documents(chunks, embeddings)
  print("Vectorstore created")
  vectorstore.save_local("/content/faiss_index")
  print("\nVector store saved to /content/faiss_index")
   # Quick test: search for something to verify it works
  test_results = vectorstore.similarity_search("main topic", k=2)
  print(f"\nTest search returned {len(test_results)} chunks — working!")
  print(f"Sample: {test_results[0].page_content[:150]}...")
  return vectorstore
#building vectorstore
vectorstore = build_vector_store(chunks)
print("embeddings completed")

```

BUILDING RAG CHAIN

```python
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
```

```python
def build_ragchain(vectorstore):

    prompt_template = """
You are a document-based QA system.

Rules:
- Use ONLY the provided context
- Combine information into a clear answer
- Do NOT copy raw chunks blindly
- Ignore irrelevant context
- If answer not found, say: "I cannot find this in the document."

Formatting:
- For definitions → short paragraph
- For advantages/disadvantages → bullet points
- Avoid repetition

Context:
{context}

Question:
{question}

Answer:
"""

    prompt = PromptTemplate(template=prompt_template)

    llm = ChatGroq(
        model="llama-3.3-70b-versatile",
        api_key=key,
        temperature=0
    )

    retriever = vectorstore.as_retriever(search_kwargs={"k": 10})

    from sentence_transformers import CrossEncoder
    cross_encoder = CrossEncoder("cross-encoder/ms-marco-MiniLM-L-6-v2")

    # 🔥 RERANK FUNCTION
    def rerank_docs(query, docs, top_k=4):
        pairs = [[query,doc.page_content] for doc in docs]
        scores = cross_encoder.predict(pairs)
        scored_docs = list(zip(docs,scores))
        scored_docs.sort(key=lambda x: x[1], reverse = True)
        return[doc for doc,_ in scored_docs[:top_k]]
    # 🔥 FINAL CONTEXT BUILDER
    def get_context(question):
        docs = retriever.invoke(question)
        reranked_docs = rerank_docs(question, docs)
        context_text = "\n\n".join(doc.page_content for doc in reranked_docs)
        context_list = [doc.page_content for doc in reranked_docs]

        return context_text, context_list


    rag_chain = (
        {
            "context": lambda q: get_context(q)[0],
            "question": RunnablePassthrough()
        }
        | prompt
        | llm
        | StrOutputParser()
    )

    return rag_chain

rag_chain = build_ragchain(vectorstore)
```

ASKING QUESTIONS AND GETTING OUTPUT

```python
def ask(question):
  print(f"Question: {question}")
  print("-" * 50)
  docs = vectorstore.similarity_search(question, k=5)
  result = rag_chain.invoke(question)
  print(f"Answer:\n{result}")
  print("\n" + "-" * 50)
  print("Sources used (chunks model read):")
  for i, doc in enumerate(docs, start=1):
      page = doc.metadata.get("page","?")
      snippet = doc.page_content[:200].replace("\n"," ")
      print("source",i,"page",page,"\nsnippet used:",snippet,"\n")
  return {"answer": result,
          "source_documents": docs
          }

```

```python
import os
from openai import OpenAI
from ragas.llms import llm_factory
from langchain_community.embeddings import HuggingFaceEmbeddings

eval_embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
# Create OpenAI-compatible client for Groq
client = OpenAI(
    api_key=os.environ["GROQ_API_KEY"],
    base_url="https://api.groq.com/openai/v1"
)

# Pass client to RAGAS
eval_llm = llm_factory(
    model="llama-3.3-70b-versatile",
    client=client
)
```

```python
from datasets import Dataset
dataset = {
    "question": eval_questions,        # ✅ NOT "questions"
    "answer": eval_answers,
    "contexts": eval_contexts,          # ✅ NOT "context"
    "ground_truth": eval_ground_truth  # optional but recommended
}
dataset = Dataset.from_dict(dataset)

```

```python
from ragas.metrics import (
    faithfulness,
    answer_relevancy,
    context_precision,
    context_recall
)
```

```python
from ragas import evaluate
from ragas.metrics import faithfulness, answer_relevancy, context_precision, context_recall
result = evaluate(
    dataset,
    metrics=[faithfulness, answer_relevancy, context_precision, context_recall],
    llm=eval_llm,
    embeddings=eval_embeddings
)

print(result)
```
